## **2. ¿Cómo funciona?**

### **2.1 Chatbot de Consulta Médica en RASA y recogida de datos en formulario**
Se implementa un chatbot diseñado con **RASA** que simula una consulta médica convencional. El chatbot interactúa con el usuario para recopilar datos esenciales como:

### **2.2 Procesamiento del Historial Clínico**
Una vez recopilada toda la información relevante del paciente, se genera un historial clínico completo en formato **CSV**. Este historial se procesa utilizando técnicas avanzadas de procesamiento del lenguaje natural. Se aplica un corpus especializado en enfermedades hematológicas para filtrar y almacenar únicamente la información que es crucial para el diagnóstico médico, como:
- Síntomas
- Medicación
- Historial personal
- Antecedentes familiares

### **2.3 Vectorización y Cálculo de Similitudes**
Los datos procesados se transforman en representaciones numéricas (**vectorización**) para permitir el cálculo de similitudes. Se realiza una comparación entre el historial clínico del paciente y cada enfermedad del corpus hematológico. Este proceso arroja porcentajes que indican la semejanza entre el historial del paciente y las posibles enfermedades identificadas.

> **Resultados:** Las enfermedades se ordenan de mayor a menor similitud, proporcionando al personal sanitario una guía inicial para el diagnóstico.

## **3. Metodología Utilizada**

El desarrollo del chatbot médico para diagnóstico previo se basa en una metodología integrada que combina técnicas avanzadas de procesamiento del lenguaje natural e inteligencia artificial:

### **3.1 Desarrollo del Chatbot con RASA**

### **3.2 Procesamiento del Historial Clínico**
Una vez recopilada la información del usuario, se genera un historial clínico estructurado en formato **CSV**. Este historial se somete a un proceso de procesamiento del lenguaje natural, que incluye tokenización, lematización y análisis de entidades para extraer información relevante. Se utiliza un corpus especializado en enfermedades hematológicas para filtrar y almacenar únicamente los datos cruciales para el diagnóstico.

### **3.3 Vectorización y Cálculo de Similitudes**

La vectorización TF-IDF es una técnica esencial en el procesamiento de texto que transforma datos textuales en vectores numéricos. TF-IDF (Term Frequency-Inverse Document Frequency) asigna pesos a las palabras en función de su frecuencia en un documento y su rareza en el corpus completo.

**1. Vectorización TF-IDF:**
En nuestro caso, TF-IDF convierte el historial clínico del paciente en vectores numéricos. Esto significa que cada palabra en el historial clínico se convierte en una característica con un valor numérico que representa su importancia relativa. Las palabras más frecuentes en el historial clínico pero raras en el corpus general tendrán pesos más altos, lo que ayuda a capturar la especificidad del caso del paciente.

**2. Cálculo de Similitudes:**
Una vez vectorizado, comparamos el historial clínico del paciente con perfiles de enfermedades almacenados en nuestro corpus hematológico. Utilizamos técnicas avanzadas de análisis de similitud, como la similitud del coseno, que mide la orientación y proximidad de los vectores en un espacio multidimensional. Cuanto más cercanos estén dos vectores en términos de similitud del coseno, mayor será la probabilidad de que el paciente pueda presentar esas enfermedades.

**3. Resultados:**
Este proceso genera porcentajes que reflejan la similitud entre el historial clínico del paciente y los perfiles de enfermedades. Estos porcentajes se ordenan de mayor a menor relevancia, proporcionando una guía clara sobre las enfermedades más probables que el paciente podría tener. Esta información es crucial para ayudar a los profesionales de la salud a realizar diagnósticos preliminares más precisos y rápidos.

- Con TF-IDF y el cálculo de similitudes, transformamos datos complejos en información útil y accesible para mejorar la atención médica

##**Paso 4: Procesamiento del Historial Clínico**

### En esta etapa, el historial clínico se prepara para el análisis mediante un proceso detallado de limpieza y estructuración.

Los datos brutos del historial clínico son transformados utilizando técnicas avanzadas de tokenización, lematización y eliminación de stopwords. Este proceso asegura que solo la información relevante y estructurada sea utilizada para identificar posibles enfermedades.

### Objetivo del Proceso

El objetivo principal es optimizar el historial clínico, convirtiéndolo en un formato procesable que permita compararlo con perfiles de enfermedades del corpus hematológico.

### Ejecución del Proceso

1. **Tokenización y Lematización:** Se dividen las palabras en unidades significativas y se reducen a su forma base para un análisis más preciso.
   
2. **Eliminación de Stopwords y Caracteres Especiales:** Se eliminan palabras sin relevancia como determinantes, conjunciones y caracteres especiales que no contribuyen al análisis diagnóstico.

Una vez completado este proceso, el historial clínico se reduce a cuatro columnas. Estas columnas contienen las categorías más relevantes para el análisis y serán utilizadas posteriormente en la búsqueda de coincidencias. La información ha sido procesada mediante tokenización y lematización, conservando únicamente los datos de valor.

```python
# Leer el archivo CSV directamente desde la ruta especificada
historial_clinico_df = pd.read_csv('/content/03_Med_Bot_ES/data/Historial_clinico.csv')


def limpiar_texto(texto):
    doc = nlp(texto.lower())
    tokens = [token.lemma_ for token in doc if token.is_alpha and token.text not in STOP_WORDS]
    return " ".join(tokens)

# Aplicar la limpieza a cada columna relevante en el DataFrame
columnas_de_interes = ['Síntomas', 'Medicación', 'Historial personal', 'Antecedentes familiares']
for columna in columnas_de_interes:
    historial_clinico_df[columna] = historial_clinico_df[columna].apply(limpiar_texto)


hc_df = historial_clinico_df #Le cambiamos el nombre para manejarlo de forma más fácil

hc_df

columnas_med = hc_df[["Historial personal", "Antecedentes familiares", "Síntomas", "Medicación"]] #Tomamos las que queremos solo. Solo estas 4 columnas van a intervenir en el análisis
columnas_med
```

```python
# Ruta del archivo en el repositorio clonado
enfermedades = "/content/03_Med_Bot_ES/enfermedades.xlsx"
pd.read_excel(enfermedades)

# Importar IPython.display para mostrar el DataFrame de manera más agradable
#el resto de bibliotecas y herramientas ya están descargadas
from IPython.display import display

# Función para limpiar el texto
def limpiar_texto(texto):
    doc = nlp(texto.lower())
    tokens = [token.lemma_ for token in doc if token.is_alpha and token.text not in STOP_WORDS]
    return " ".join(tokens)

# Leer el archivo Excel desde la ruta especificada
file_path = "/content/03_Med_Bot_ES/enfermedades.xlsx"
df = pd.read_excel(file_path)

# Extraer la columna 'Tratamiento' antes de limpiarla
tratamiento_df = df[['Enfermedad', 'Tratamiento']].copy()

# Aplicar la limpieza a cada columna relevante en el DataFrame (Causas, Síntomas y Tratamiento)
columnas_a_limpiar = ['Causas', 'Síntomas', 'Tratamiento']
for columna in columnas_a_limpiar:
    df[columna] = df[columna].apply(limpiar_texto)

# Mostrar el DataFrame procesado modo tabla
display(df)
```

### **Paso 6: Búsqueda de Coincidencias**

¡Excelente! Ahora procedemos a buscar coincidencias entre el historial clínico y el corpus de enfermedades mediante dos procesos fundamentales:

**Primero: Vectorización**
Transformamos las palabras de nuestros textos en números para facilitar la comparación de similitudes entre ellos.

**Segundo: Similitud de Cosenos**

Utilizamos la similitud de cosenos, una métrica que evalúa la proximidad entre dos vectores en un espacio multidimensional. Esta técnica nos permite medir qué tan similar es el historial clínico de un paciente a cada una de las enfermedades contenidas en nuestro corpus.

**Limitaciones del Modelo por la Similitud de cosenos:**
- Sensible a la calidad y cantidad de datos del historial clínico y del corpus de enfermedades.
- No considera contexto clínico específico ni información temporal de los síntomas.
- Puede no capturar similitudes sutiles o características únicas de ciertas enfermedades.

Gracias a esta medida, podemos identificar potenciales enfermedades del paciente basándonos únicamente en el análisis de palabras y cálculos matemáticos.

Por favor, ejecuta la siguiente celda para obtener los resultados. Antes de descargar los archivos finales en formato Excel, podrás revisar los resultados en formato de texto y visualizar gráficas como la matriz de similitud de cosenos, gráficos de barras y una nube de palabras.

---

**Aviso Legal:**

Ten en cuenta que los resultados presentados a continuación son generados automáticamente. La información proporcionada debe ser complementada con el criterio clínico y no sustituye el consejo médico profesional. Las decisiones relacionadas con el cuidado del paciente deben ser tomadas por un profesional sanitario autorizado.
```python
# Concatenar las columnas del historial clínico en un solo texto por cada fila
historial_texts = columnas_med.apply(lambda row: ' '.join(row.dropna().values.astype(str)), axis=1)

# Concatenar las columnas del corpus de enfermedades (causas, síntomas, tratamientos)
enfermedades_texts = df['Causas'] + ' ' + df['Síntomas'] + ' ' + df['Tratamiento']

# Vectorización TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Vectorizar los textos del historial clínico y del corpus de enfermedades
tfidf_historial = tfidf_vectorizer.fit_transform(historial_texts)
tfidf_enfermedades = tfidf_vectorizer.transform(enfermedades_texts)

# Calcular la similitud del coseno
similitudes = cosine_similarity(tfidf_historial, tfidf_enfermedades)

# Verificar la longitud de los nombres de enfermedades y las dimensiones de la matriz de similitudes
print("Longitud de nombres de enfermedades:", len(df))
print("Dimensiones de la matriz de similitudes:", similitudes.shape)

# Asegurarse de que las dimensiones coincidan
assert len(df) == similitudes.shape[1], "Mismatch in dimensions"

# Convertir la matriz de similitudes en un DataFrame con nombres de enfermedades
similitudes_df = pd.DataFrame(similitudes, columns=df['Enfermedad'])

# Mostrar las enfermedades más probables para el primer historial clínico
top_similitudes = similitudes_df.iloc[0].sort_values(ascending=False)
print("Top enfermedades probables:", top_similitudes.head(10))

# Mostrar el DataFrame procesado
display(similitudes_df)


# Crear un DataFrame con los resultados ordenados
resultados_ordenados = pd.DataFrame({'Enfermedad': top_similitudes.index, 'Similitud': top_similitudes.values})

# Guardar el DataFrame en un archivo Excel
resultados_ordenados.to_excel('posibles_enfermedades.xlsx', index=False)

# Visualizar la matriz de similitud como un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(similitudes_df, cmap="YlGnBu")
plt.title("Matriz de Similitud de Coseno entre Historiales Clínicos y Enfermedades")
plt.xlabel("Enfermedades")
plt.ylabel("Historiales Clínicos")
plt.show()
# Guardar el DataFrame procesado en un nuevo archivo Excel
#similitudes_df.to_excel('similitudes_enfermedades.xlsx', index=False)

from wordcloud import WordCloud

# Obtener las enfermedades más probables para el primer historial clínico
top_similitudes = similitudes_df.iloc[0].sort_values(ascending=False)

# Generar una lista de las palabras comunes más relevantes de las enfermedades
palabras_comunes = ' '.join(top_similitudes.index)

# Crear una nube de palabras
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(palabras_comunes)

# Visualizar la nube de palabras
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("Nube de Palabras de Enfermedades Identificadas por Similitud de Coseno")
plt.axis('off')
plt.show()


# Gráfico de barras de las enfermedades más probables
plt.figure(figsize=(10, 6))
top_similitudes.head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Enfermedades Probables")
plt.xlabel("Enfermedad")
plt.ylabel("Similitud de Coseno")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Gráfico de distribución de similitudes
plt.figure(figsize=(8, 6))
sns.histplot(similitudes.flatten(), kde=True, color='salmon')
plt.title("Distribución de Similitudes de Coseno")
plt.xlabel("Similitud de Coseno")
plt.ylabel("Frecuencia")
plt.show()
```
